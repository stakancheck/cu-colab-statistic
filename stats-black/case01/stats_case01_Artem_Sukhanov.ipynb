```python
palette = {
    'light': '#FAFAFA',
    'black': '#141414',
    'gray-light': '#E6E6E6',
    'sub-gray': '#969696',
    'green': '#009B40',
    'electric_blue': '#3044FF',
    'ambitious_red': '#3044FF',
    'legent_yellow': '#FFDD2D',
    'bright_fuchsia': '#DB45E8',
}

def stylize(plt, fig, ax):
    # Style settings for CU style
    plt.rcParams["font.family"] = 'Graphik LCG'
    # Set outline colors
    plt.rcParams['axes.edgecolor'] = palette['sub-gray']  # Color of the axes edge
    plt.rcParams['xtick.color'] = palette['sub-gray']     # Color of the x-tick labels
    plt.rcParams['ytick.color'] = palette['sub-gray']     # Color of the y-tick labels
    plt.rcParams['text.color'] = palette['black']      # Color of the text
    # Set font color
    plt.rcParams['axes.labelcolor'] = palette['black']  # Color of the x and y labels
    plt.rcParams['axes.titlecolor'] = palette['black']  # Color of the title
    fig.set_facecolor(palette['light'])  # Цвет фона фигуры
    ax.set_facecolor(palette['light'])  
    # Цвет фона осей
    plt.xticks(fontweight='semibold')
    plt.yticks(fontweight='semibold')
```

```python
# Разделим presence_russia на обычные и франчайзинговые точки
def process_presence_russia(value):
    if isinstance(value, str):
        if 'и' in str(value):
            parts = str(value).split(' и ')
            return int(parts[0]) + int(parts[1].split()[0])
        return int(value)
    elif isinstance(value, float):
        return value
    return None

def get_franchise_presence_russia(value):
    if isinstance(value, str):
        if 'и' in str(value):
            parts = str(value).split(' и ')
            return int(parts[1].split()[0])
        return 0
    elif isinstance(value, float):
        return 0
    return 0

df['presence_world'] = df['presence_world'].fillna(0)
df['presence_russia_franchise'] = df['presence_russia'].apply(get_franchise_presence_russia)
df['presence_russia_combined'] = df['presence_russia'].apply(process_presence_russia)
df.head(20)
```

```python
categories = {
    'Локальные ритейлеры': (1, 10),
    'Региональные ритейлеры': (11, 50),
    'Национальные ритейлеры': (51, 500),
    'Крупные национальные ритейлеры': (501, 2000),
    'Международные ритейлеры': (2001, 5000),
    'Глобальные ритейлеры': (5001, float('inf'))
}

def categorize_retailer(points):
    for category, (low, high) in categories.items():
        if low <= points <= high:
            return category
    return None
    
df['category'] = df['presence_russia_combined'].apply(categorize_retailer)

# Подсчет количества ритейлеров в каждой категории
category_counts = df['category'].value_counts().reindex(categories.keys(), fill_value=0)

fig, ax = plt.subplots(figsize=(10, 6))

stylize(plt, fig, ax)

ax.bar(category_counts.index, category_counts.values, color=palette['electric_blue'])

ax.set_xlabel('Категории ритейлеров')
ax.set_ylabel('Количество ритейлеров')
# ax.set_title('Распределение ритейлеров по категориям')
ax.grid(axis='y')

# Поворот меток на оси X для лучшей читаемости
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.savefig('retailers_per_category.png', dpi=300)
plt.show()
```

```python
# Подсчет количества регионов присутствия для каждой категории
df['presence_regions_count'] = df['presence_regions'].apply(lambda x: len(str(x).split(';')) if isinstance(x, str) else 0)

# Группировка данных по категориям и подсчет количества ритейлеров в каждом регионе
region_distribution = df.groupby(['category', 'presence_regions_count']).size().unstack(fill_value=0)

fig, ax = plt.subplots(figsize=(12, 8))

region_distribution.T.plot(kind='bar', stacked=True, ax=ax, colormap='tab20')

ax.set_xlabel('Количество регионов присутствия')
ax.set_ylabel('Количество ритейлеров')
ax.set_title('Распределение ритейлеров по регионам в зависимости от категории')
ax.legend(title='Категории ритейлеров', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.savefig('region_distribution_per_category.png', dpi=300)
plt.show()
```

```python
# Task 10
df_russia = df[df['presence_russia_combined'] > 0]

df_regions = df_russia.assign(presence_regions=df['presence_regions'].str.split(';')).explode('presence_regions')

top10_regions = df_regions['presence_regions'].value_counts().head(10).index

df_regions = df_regions.groupby(['category', 'presence_regions']).size().reset_index(name='count')

def process_regions_to_top(row):
    if row in top10_regions:
        return row
    return 'Другие'

df_regions['presence_regions'] = df_regions['presence_regions'].apply(process_regions_to_top)


pivot_table_regions = df_regions.pivot_table(
    index='presence_regions', 
    columns='category', 
    values='count', 
    aggfunc='sum', 
    fill_value=0
)

category_order = ['Локальные ритейлеры', 'Региональные ритейлеры', 'Национальные ритейлеры', 'Крупные национальные ритейлеры', 'Международные ритейлеры', 'Глобальные ритейлеры']
pivot_table_regions = pivot_table_regions[category_order]


colors = [
    '#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd',
    '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf', '#c7c7c7'
]

fig, ax = plt.subplots(figsize=(12, 8))

stylize(plt, fig, ax)

# Построение сегментированной столбчатой диаграммы
pivot_table_regions.T.plot(kind='bar', stacked=True, ax=ax, color=colors)

ax.set_xlabel(None)
ax.set_ylabel('Количество ритейлеров')
ax.set_title('Распределение ритейлеров по регионам в зависимости от категории')
handles, labels = ax.get_legend_handles_labels()
sorted_handles_labels = sorted(zip(handles, labels), key=lambda x: list(top10_regions).index(x[1]) if x[1] in top10_regions else len(top10_regions))
sorted_handles, sorted_labels = zip(*sorted_handles_labels)
ax.legend(sorted_handles, sorted_labels, title='Регионы', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.savefig('segmented_barchart_regions_per_category.png', dpi=300)

```
