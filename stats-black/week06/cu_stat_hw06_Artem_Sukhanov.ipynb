# Задача 1. Покажи поведение выборки из распределения



Узнай, насколько точно статистика, найденная по выборке, приближает аналогичную характеристику в генеральной совокупности. Сделать это нужно для простой случайной выборки размера 1, 10, 500, для стратифицированной двумя разными способами случайной выборки размера 10 и 500; для среднего, 1-й квартили и максимума.

На всех гистограммах в этой задаче отметь среднее в генеральной совокупности. Это нужно, чтобы по графику было видно, насколько точно среднее по выборке приближает среднее в генеральной совокупности.

В твоём распоряжении датасет [time_mobile_app.csv](https://raw.githubusercontent.com/zalig/cu-datasets/main/06_time_mobile_app.csv). В нём есть информация обо всех пользователях мобильного приложения (генеральная совокупность).

Для всех пунктов, где потребуется, количество выборок $N=10000$.

T — время, проведённое клиентом в мобильном приложении, в часах.


## 1. (**blue, баллы: 1.5**) Среди всех пользователей:
- Найди среднее время, проводимое клиентами в мобильном приложении, и смещённое стандартное отклонение.
- Найди распределение T и покажи результат на гистограмме.


**Решение**

```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/zalig/cu-datasets/main/06_time_mobile_app.csv')
df.head()
```

```python
import matplotlib.pyplot as plt

plt.hist(df['T'], bins=50, color='blue', alpha=0.7)
plt.axvline(df['T'].mean(), label='Среднее время', color='red', linestyle='dashed', linewidth=1)

plt.title('Время, проведенное клиентами в мобильном приложении')
plt.xlabel('Время, часы')
plt.ylabel('Количество клиентов')
plt.show()

# Смещенное стандартное отклонение
std = df['T'].std()
print('Смещенное стандартное отклонение:', float(std))

# Среднее значение
mean = df['T'].mean()
print('Среднее значение:', float(mean))
```

## 2. (**blue, баллы: 1.5**) Построй $N$ простых случайных выборок размера 1 клиент. Среди них:
  - Найди среднее T и его смещённое стандартное отклонение.
  - Найди распределение T и покажи результат на гистограмме.


**Решение**

```python
import numpy as np

N = 10000
sample_means = np.zeros(N)

for i in range(N):
    sample = df['T'].sample(1, replace=False)
    sample_means[i] = sample.mean()
    
plt.hist(sample_means, bins=50, color='blue', alpha=0.7)
plt.axvline(sample_means.mean(), label='Среднее время', color='red', linestyle='dashed', linewidth=1)

plt.title('Время, проведенное клиентами в мобильном приложении')
plt.xlabel('Время, часы')
plt.ylabel('Количество клиентов')
plt.show()

# Смещенное стандартное отклонение
std_srs_1 = sample_means.std()
print('Смещенное стандартное отклонение:', float(std_srs_1))

# Среднее значение
mean_srs_1 = sample_means.mean()
print('Среднее значение:', float(mean_srs_1))
```

## 8. (**black, баллы: 2**) Построй $N$ выборок, стратифицированных по новизне модели мобильного устройства, посчитай по каждой такой выборке среднее $\bar{T}$ и среди них:
  - Найди среднее $\bar{T}$ и его смещённое стандартное отклонение.
  - Найди распределение $\bar{T}$ и покажи результат на гистограмме.
  - Сравни с предыдущим при размере выборки 10 клиентов и 500 клиентов.


**Решение**

```python
df['Model'].value_counts()
```

```python
means = []
stds = []

mean_t = df['T'].mean()

def stratified_sample(df, strata_col, frac):
    groups = df.groupby(strata_col)
    sample = pd.DataFrame()

    for _, group in groups:
        stratum_sample = group.sample(frac=frac, replace=False)
        sample = pd.concat([sample, stratum_sample])

    return sample

n_all = len(df)
sample_size = 10

strat_props = []

for _ in range(N):
    df_sample = stratified_sample(df, 'Model', sample_size / n_all)
    strat_prop = df_sample['T'].mean()
    strat_props.append(strat_prop)

mean_str_model_10 = pd.Series(strat_props).mean()
std_str_model_10 = pd.Series(strat_props).std()

means.append(mean_str_model_10)
stds.append(std_str_model_10)

print('Среднее значение:', mean_str_model_10)
print('Смещенное стандартное отклонение:', std_str_model_10)

plt.hist(strat_props, label='Средние', alpha=0.6, bins=50)
plt.axvline(mean_t, label='Среднее время', color='red')

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.title('Время, проведенное клиентами в мобильном приложении по моделям (выборка 10)')
plt.xlabel('Время, часы')
plt.ylabel('Количество клиентов')
plt.show()
```

```python
means = []
stds = []

n_all = len(df)
sample_size = 500

strat_props = []

for _ in range(N):
    df_sample = stratified_sample(df, 'Model', sample_size / n_all)
    strat_prop = df_sample['T'].mean()
    strat_props.append(strat_prop)

mean_str_model_500 = pd.Series(strat_props).mean()
std_str_model_500 = pd.Series(strat_props).std()

means.append(mean_str_model_500)
stds.append(std_str_model_500)

print('Среднее значение:', mean_str_model_500)
print('Смещенное стандартное отклонение:', std_str_model_500)

plt.hist(strat_props, label='Средние', alpha=0.6, bins=50)
plt.axvline(mean_t, label='Среднее время', color='red')

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.title('Время, проведенное клиентами в мобильном приложении по моделям (выборка 500)')
plt.xlabel('Время, часы')
plt.ylabel('Количество клиентов')
plt.show()
```

**Сравнение:**

Среднее значение времени, проведенного клиентами в мобильном приложении, при стратифицированной выборке по моделям приближается к среднему значению времени по генеральной совокупности. При увеличении размера выборки точность приближения к среднему генеральной совокупности растет. При этом смещенное стандартное отклонение уменьшается. Если сравнивать данные последние рассчеты с предыдущими соотвественно для 10 и 500 выборок, то можно сделать вывод, что стратифицированные выборки дают приближенные результаты по среднему и смещенному стандартному отклонению.
