### Преобразование describe для удобного чтения

```python
# Format the describe output for better readability

def format_large_numbers(x):
    if x >= 1_000_000_000:
        return f"{x / 1_000_000_000:,.0f}B"
    elif x >= 1_000_000:
        return f"{x / 1_000_000:,.0f}M"
    elif x >= 1_000:
        return f"{x / 1_000:,.2f}K"
    else:
        return f"{x:.2f}"
    
# Convert data in columns in new format
formatted_output = describe_output.style.format(format_large_numbers)
formatted_output
```

### Сегментрированный bar chart для уровня образования в топ-10 странах

Выведем количество людей с разным уровнем образования в топ-10 странах. Это позволит нам понять, как уровень образования варьируется в разных странах.

```python
# Plot the distribution of education level in the top-10 countries

# Filter the dataframe to include only the top 10 countries
df_top_countries = df[df['Country'].isin(top_countries)]

# Drop rows with NaN values in 'EdLevel' and 'Country'
df_top_countries = df_top_countries.dropna(subset=['EdLevel', 'Country'])

# Plot the stacked bar chart
plt.figure(figsize=(14, 8))
sns.histplot(data=df_top_countries, x='Country', hue='EdLevel', multiple='stack', palette='viridis', shrink=0.8)
plt.title('Distribution of Education Level in the Top-10 Countries')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Сгементированная столбачатая диаграмма для уровня образования в топ-10 странах

Чтобы лучше понять распределение разных степеней образования в топ-10 странах, построим столбчатую диаграмму отношения, где каждый столбец будет представлять одну страну, а высота столбца будет показывать долю каждого уровня образования в этой стране.

```python
# Filter the dataframe to include only the top 10 countries
df_top_countries = df[df['Country'].isin(top_countries)]

# Drop rows with NaN values in 'EdLevel' and 'Country'
df_top_countries = df_top_countries.dropna(subset=['EdLevel', 'Country'])

# Calculate the proportion of each education level within each country
df_proportion = df_top_countries.groupby(['Country', 'EdLevel']).size().reset_index(name='Count')
df_proportion['Proportion'] = df_proportion.groupby('Country')['Count'].transform(lambda x: x / x.sum())

# Plot the stacked bar chart
plt.figure(figsize=(14, 8))
hist = sns.histplot(data=df_proportion, x='Country', weights='Proportion', hue='EdLevel', multiple='stack', palette='viridis', shrink=0.8)
plt.title('Proportion of Education Level in the Top-10 Countries')
plt.xlabel('Country')
plt.ylabel('Proportion')
sns.move_legend(hist, "upper left", bbox_to_anchor=(1, 1))
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### График зависимости заработной платы от опыта работы в топ-10 странах

```python
# Filter the dataframe to include only the top 10 countries
df_top_countries = df[df['Country'].isin(top_countries)]

# Replace 'More than 50 years' with '50+' (for sorting)
df_top_countries['YearsCodePro'] = df_top_countries['YearsCodePro'].replace('More than 50 years', '51')

# Replace 'Less than 1 year' with '0'
df_top_countries['YearsCodePro'] = df_top_countries['YearsCodePro'].replace('Less than 1 year', '0')

# Filter NaN and No values
df_top_countries['YearsCodePro'] = df_top_countries['YearsCodePro'].replace({'NaN': np.nan, 'No': np.nan})
df_top_countries['YearsCodePro'].unique()
```
