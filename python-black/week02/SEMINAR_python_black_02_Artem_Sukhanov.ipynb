## Задача 1






>**Кодировка** — это способ представления символов в виде байтов. Существует несколько типов кодировок:
  - **ASCII.** 7-битная, поддерживает только английский язык.
  - **UTF-8.** Самая популярная, поддерживает все символы Юникода. Совместима с ASCII и использует переменное количество байтов (от 1 до 4).
  - **Windows-1251.** Используется для кириллицы в старых системах.
  - **KOI8-R.** Кодировка, которая когда-то давно применялась в Unix-системах и на многих сайтах в интернете.
>  
>Неправильное определение кодировки может привести к искажению текста. Это особенно критично в компании, предоставляющей клиентские услуги. Поэтому важно убедиться, что текстовые данные клиентов закодированы корректно.

1. Создай файл `detect_encoding.py`.

2. Напиши программу, которая с помощью библиотеки `chardet` определяет кодировку строки текста и выводит название кодировки на экран.

3. Запусти программу с различными текстовыми строками и убедись, что она правильно определяет их кодировку.






  Примеры строк для тестирования:
  - `Hello, how are you?` (ASCII)
  - `Привет, как дела?` (UTF-8)
  - `こんにちは、元気ですか？` (UTF-8)
  - `¡Hola! ¿Cómo estás?` (UTF-8)


## Задача 2




>**Перекодировка** — это процесс преобразования текста из одной кодировки в другую. Например, текст часто перекодируют из Windows-1251 в UTF-8, чтобы работать с ним на современных системах. А для работы с юридическими документами требуется кодировка Windows-1251.

1. Создай файл `convert_to_win1251.py`.
2. Напиши программу, которая перекодирует введённый текст в Windows-1251 и выведи его на экран.
3. Убедись, что программа корректно обрабатывает строки на разных языках, включая русский.
  


  Примеры строк для тестирования:
  - `Этот документ требует согласования CEO.` (UTF-8)
  - `Документ должен быть сохранён.` (UTF-8)
  - `Здесь используется кодировка Windows-1251.` (UTF-8)

## Задача 3

Компания получает сообщения от клиентов на разных языках. Перед тем, как сохранить сообщения в базу данных, их нужно привести к единой кодировке UTF-8.

1. Создай файл `standardize_encoding.py`.
2. Напиши программу, которая принимает строку текста, определяет её кодировку и перекодирует её в UTF-8.
3. Проверь программу, введя текстовые строки на разных языках.



  Примеры строк для тестирования:
  - `Bonjour, comment ça va?` (ISO-8859-1)
  - `Привет! Когда мы можем встретиться?` (UTF-8)
  - `Спасибо за вашу помощь.` (UTF-8)
  - `Auf Wiedersehen!` (ASCII)

## Задача 4

Ты работаешь в отделе продаж, и тебе нужно убедиться, что данные от компаний-партнёров из разных стран приведены к стандартной кодировке.

1. Создай файл `sales_data_analysis.py`.
2. Напиши программу, которая принимает три строки с информацией о продажах (название товара, количество, цена),
определяет их кодировку и конвертирует в UTF-8.
3. Убедись, что программа корректно обрабатывает данные на разных языках.



  Примеры строк для тестирования:
  - `Товар: Кофе; Количество: 10; Цена: 500 рублей` (UTF-8)
  - `Product: Laptop; Quantity: 5; Price: $1200` (ASCII)
  - `Producto: Teléfono; Cantidad: 20; Precio: 300 EUR` (MacRoman)

## Задача 5

Для компании маркетинговых исследований нужно написать программу, которая:
- принимает текст на любом языке,
- определяет его кодировку,
- подсчитывает количество уникальных слов в этом тексте и выводит их на экран.

Для этого:

1. Создай файл `count_unique_words.py`.
2. Напиши программу, которая читает текст, определяет его кодировку с помощью библиотеки `chardet` и перекодирует текст в UTF-8 (если требуется).
3. Программа должна вывести все уникальные слова (каждое с новой строки) и количество уникальных слов.
4. Убедись, что программа правильно обрабатывает тексты на разных языках.

Пример вывода:

```text
Слова:
time
on
and
met
Highly
expectations.
delivered
The
was
all
product
recommended!
Количество уникальных слов: 12
```

Примеры строк для тестирования:

- `Отличное обслуживание и быстрый сервис. Продукт очень понравился!`
- `The product was delivered on time and met all expectations. Highly recommended!`
- `El servicio al cliente fue excelente, y el producto llegó en perfectas condiciones.`
- `Продажи в январе выросли на 15%. Объём продаж составил 200 единиц.`
- `Sales in January increased by 15%. The total sales volume was 200 units.`
- `Der Kundenservice war ausgezeichnet, und das Produkt wurde pünktlich geliefert.`

## Задача 6

Иногда в текстах от клиентов встречаются неправильно написанные слова.Необходимо автоматизировать процесс их исправления с помощью заранее подготовленного словаря исправлений.


1. В файле `auto_correct.py` напиши программу, которая читает строку текста, определяет её кодировку и исправляет ошибки с использованием словаря.
2. Убедись, что программа корректно исправляет типичные ошибки, такие как `превет`, `спсибо`, `кагда`, и возвращает правильный текст.

Примеры строк для тестирования:

- `Дорогой клиент, преветствуем вас на борту!`

  Должно исправиться на «Дорогой клиент, приветствуем вас на борту!».

- `Спсибо за оформлние заказа, кагда ваш заказ будет готов, мы сообщим вам.`

  Должно исправиться на «Спасибо за оформление заказа. Когда ваш заказ будет готов, мы сообщим вам».
- `Добрый день! Напоминаем, что ваше время для визита в офис назначено на сегодна.`

  Должно исправиться на «Добрый день! Напоминаем, что ваше время для визита в офис назначено на сегодня».

- `Отчёт по продажам за квартал: продажа возросла на 15%. Общий обьём продаж составил 500 единиц.`

  Должно исправиться на «Отчёт по продажам за квартал: продажа возросла на 15%. Общий объём продаж составил 500 единиц».
