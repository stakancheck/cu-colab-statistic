### Задача 1


1. Создай файл `create_thumbnails.py`.
2. Напиши функцию `create_thumbnails()`, которая открывает каждую фотографию из директории `photos_for_processing`, уменьшает её до миниатюры (размером 150x150 пикселей) и сохраняет в **новую директорию** `thumbnails` в корне проекта с префиксом `thumb_` перед названием файла.

  >**Подсказка.**
  Для работы с файлами в операционной системе через Python используй пакет `os` из стандартной библиотеки Python.
  - Чтобы обработать все изображения из директории, запусти цикл по её содержимому. Содержимое директории можно получить с помощью метода `.listdir()` пакета `os`. <br>
  [Документация по методу `.listdir()`](https://docs.python.org/3/library/os.html#os.listdir)
  - Используй метод `.makedirs()` пакета `os`, чтобы создать новую директорию. Учти случай, когда директория уже создана. <br>
  [Документация по методу `.makedirs()`](https://docs.python.org/3/library/os.html#os.makedirs)
  - Чтобы объединить строки в путь к файлу для его сохранения, используй метод `.join()` из модуля `path` пакета `os`. <br>
  [Документация по методу `.join()`](https://docs.python.org/3/library/os.path.html#os.path.join)

3. Сделай программу универсальной — добавь возможность задать размер миниатюр. Для этого нужно запросить данные у пользователя и добавить функции аргумент `thumbnail_size` со значением по умолчанию.
4. Вызови написанную функцию и проверь её работу.
