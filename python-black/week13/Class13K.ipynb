## Подготовка

1. Поделитесь на&nbsp;команды по два человека: один участник пары хостит сервер, другой подключается как клиент.

2. Организуйте сеть Wi-Fi: один раздаёт Wi-Fi с&nbsp;телефона, второй подключается к этой сети.

3. Настройте хост сервера. Хост использует скрипт get_loc_ip.py для получения локального IP-адреса в&nbsp;сети Wi-Fi.

4. Замените IP-адрес в&nbsp;серверной и&nbsp;клиентской частях кода на&nbsp;локальный IP, полученный из&nbsp;get_loc_ip.py.

5. Обсудите механику, протестируйте работу локально и&nbsp;начните игру!


## Рекомендации

1. Тестируйте код локально:

  - каждый участник запускает и&nbsp;сервер, и&nbsp;клиент на&nbsp;своём устройстве;
  - не&nbsp;передавайте сервер одному участнику, а клиент — другому. Это может вызвать технические проблемы.

2. Обсуждайте идеи:
  - договоритесь, как будете решать задачу;

3. Планируйте задачи:
  - один из&nbsp;участников может заниматься серверной частью, другой — клиентской.
  - по мере выполнения меняйтесь ролями, чтобы каждый освоил обе стороны.

4. Обсудите результаты после окончания:
   - что получилось;
   - какие трудности возникли;
   - как можно улучшить процесс.


## Задачи


### Задача 1
  
Модифицируй клиентскую часть игры «Камень, ножницы, бумага». Добавь отображение хода соперника после каждого раунда.

#### Инструкция. Клиентская часть (`rps_client.py`)

**Отображение ходов в&nbsp;сообщении `result`**
1. Модифицируй обработку сообщения о результате, чтобы отображать ход каждого игрока. Убедись, что клиент правильно отображает ходы:
     - для игрока 1: отображает ход как «Игрок 1», ход соперника — «Игрок 2»;
     - для игрока 2: отображает ход как «Игрок 2», ход соперника — «Игрок 1».
2. После отображения ходов добавь сообщение об исходе раунда: ничья, победа или поражение.




>**Подсказка.** Тестируй локально, чтобы убедиться в&nbsp;корректности отображения.

**Пример.** После завершения раунда игрок получит сообщение с&nbsp;результатом и&nbsp;ходами, которые сделали он и&nbsp;соперник.

Ожидаемый вывод при&nbsp;победе:
  ```
  Игрок 1 выбрал: камень.
  Игрок 2 выбрал: ножницы.
  Ваш ход: камень, ход соперника: ножницы.
  Вы победили!
  ```

При ничьей:
  ```
  Ничья!
  ```
При поражении:
  ```
  Вы проиграли.
  ```

### Задача 2

Доработай клиентскую часть игры, чтобы она проверяла корректность введённых данных перед отправкой на&nbsp;сервер. Проверка предотвратит отправку неверных ходов и&nbsp;улучшит пользовательский интерфейс.

#### Инструкция. Клиентская часть (`rps_client.py`)

**Обработка сообщения `your_move`**

Добавь цикл для&nbsp;проверки хода игрока. Если игрок вводит неверный ход, программа запрашивает ввод заново, пока он не&nbsp;введёт «камень», «бумага» или&nbsp;«ножницы».

**Проверка ввода**
1. Убедись, что вводимые данные не&nbsp;чувствительны к регистру. Преобразуй их в&nbsp;нижний регистр перед отправкой на&nbsp;сервер.
2. Если введённый ход некорректен, выведи сообщение: `Неверный ход. Пожалуйста, введите 'камень', 'бумага' или 'ножницы'`.



>**Подсказка**. Проверь, как программа реагирует на&nbsp;некорректный ввод. Обсуди с&nbsp;напарником, как лучше организовать проверку.

**Пример.** При запуске и&nbsp;подключении к&nbsp;серверу игрок получает сообщение с&nbsp;просьбой ввести ход. Если он вводит неверное значение, программа запрашивает ввод до&nbsp;тех пор, пока не&nbsp;получит допустимый ход.

Ожидаемый вывод (при неверном ходе):
```
Введите ход (камень, бумага или ножницы): тест.
Неверный ход. Пожалуйста, введите 'камень', 'бумага' или 'ножницы'.
Введите ход (камень, бумага или ножницы): камень.
```



### Задача 3

Модифицируй серверную и&nbsp;клиентскую части, чтобы добавить возможность повторной игры. После окончания игры сервер спросит обоих игроков, хотят ли они сыграть ещё раз. Если оба согласны, игра перезапускается. Если один из&nbsp;игроков отказывается, соединение завершается.

#### Инструкция. Серверная часть (`rps_server.py`)

**Метод `ask_for_rematch`**

Добавь этот метод в&nbsp;класс `RockPaperScissorsGame`. Он отправляет запрос на&nbsp;переигровку обоим игрокам, собирает их ответы и&nbsp;решает: перезапустить или завершить игру.
   1. После отправки результата игры в&nbsp;методе `determine_winner` вызови `await self.ask_for_rematch()`, чтобы запросить переигровку.
   2. Что произойдёт потом:
     - метод отправит сообщение с&nbsp;запросом обоим игрокам;
     - получит ответы от каждого игрока (через `receive_response`);
     - если оба ответа «да», вызовет `reset_game()` и&nbsp;запустит новый раунд с&nbsp;помощью `start_game`;
     - если хотя бы один ответ «нет», отправит игрокам сообщение об окончании игры.

**Метод `receive_response`**

Создай метод в&nbsp;`RockPaperScissorsGame`, чтобы получать ответ от&nbsp;игрока на&nbsp;запрос переигровки.
1. Получи ответ от игрока в формате JSON.
2. Проверь, равен ли ответ `'да'`, и&nbsp;верни `True` для положительного ответа и&nbsp;`False` для отрицательного и при&nbsp;разрыве соединения.

**Метод `reset_game`**

Создай этот метод в&nbsp;`RockPaperScissorsGame`, чтобы сбрасывать состояние игры перед новым раундом.

1. В `reset_game` очисти ходы игроков.
2. Сбрось флаг `game_over` в&nbsp;`False`, чтобы позволить начать новый раунд.

#### Инструкция. Клиентская часть (`rps_client.py`)

**Обработка сообщения `rematch`**
1. Добавь обработку этого сообщения в&nbsp;основном игровом цикле в&nbsp;методе `play`.
2. При получении сообщения `rematch`:
     - выведи игроку запрос, хочет ли он сыграть снова;
     - введи проверку, чтобы игрок мог ввести только «да» или «нет». Если ответ неверный, повтори запрос;
     - отправь выбранный ответ на&nbsp;сервер.

**Обработка сообщения `end`**
1. Добавь обработку этого сообщения для&nbsp;завершения игры.
2. При получении `end` выведи сообщение об окончании игры и&nbsp;заверши соединение.



>**Подсказка.** Проверь сценарии, когда оба согласны и&nbsp;когда один отказывается. Убедись, что сервер корректно обрабатывает ответы.

**Пример.** После окончания игры игрокам предложат сыграть снова.

Ожидаемый вывод:
```
Игрок 1 выбрал: камень.
Игрок 2 выбрал: ножницы.
Ваш ход: камень, ход соперника: ножницы.
Вы победили!
Хотите сыграть снова? (да/нет) да.
```

Если хотя бы один игрок откажется:
```
Хотите сыграть снова? (да/нет) нет.
Игра завершена.
```
