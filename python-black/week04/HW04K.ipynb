## Задача 2

Диана узнала, что ребята из группы хотят сделать подарок, и ей не терпится узнать, что это. Также она услышала, что идеи для подарка зашифрованы в изображениях, но точно не известно, в каких.

Напиши программу, которая обнаружит изображения со скрытым текстом по ключевому слову «подарок».

### Шаг 1
  - Настрой окружение в проекте.
  - Создай новый файл `image_link_detector.py` в своём проекте. Установи библиотеку `Pillow`, если она ещё не установлена, чтобы работать с изображениями.
  
  ```bash
  pip install pillow
  ```
    
### Шаг 2
- Напиши функцию `extract_data_from_image(...)`, которая ищет данные в изображении. Считай, что информация закодирована методом наименьшего значащего бита (LSB).

  Функция принимает аргумент `image_path` — это путь к файлу изображения, из которого необходимо извлечь данные. Возвращает декодированные из изображения данные в виде строки.

- Чтобы найти текст, напиши функцию `find_key_text(extracted_data)`, которая ищет ключевое слово «подарок» и выводит строку с ним. Функция принимает аргумент `extracted_data` — извлечённые из изображения данные в виде строки. Возвращает список строк с найденным ключевым словом.

    
    
### Шаг 3
Напиши функцию `analyze_images(image_paths)` основной программы, которая:
- Принимает список изображений, которые необходимо проанализировать.
- Извлекает из каждого изображения скрытые данные.
- Ищет слово «подарок» и выводит строки с этим ключевым словом.
- Выводит на экран список файлов, в которых обнаружено ключевое слово, в формате: <br>
 `Ключевое слово найдено в файлах ..., ... `
 <br> Если в изображениях ключевое слово не найдено, выводит сообщение:
 <br> `Ключевое слово не найдено ни в одном из файлов`.
