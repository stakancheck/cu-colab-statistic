2. Каждому десятому клиенту — скидка 10% на следующую покупку.
3. Каждому пятому клиенту — подарочный купон на сумму 200 рублей.

Клиент может получить только один вид бонуса, причём наиболее редкий из доступных.

Напиши программу, которая отслеживает число обслуженных клиентов, суммирует выручку за день, а также регистрирует количество выданных бонусных предложений каждого типа. Программа прекращает работу по команде `Конец рабочего дня` и выводит 5 чисел через пробел:

- количество клиентов за сутки;
- выручку;
- количество скидок 10%;
- количество купонов на 200 рублей;
- количество плюшевых медведей.

````python
# Твой код здесь

# Задача 5
clients = 0
revenue = 0
discount_10_percent = 0
coupon_200_rub = 0
teddy_bear = 0

while True:
    input_value = input("Введите выручку или 'Конец рабочего дня': ")
    if input_value.lower() == "конец рабочего дня":
        break
    
    revenue += float(input_value)
    clients += 1
    
    if clients % 15 == 0:
        teddy_bear += 1
    elif clients % 10 == 0:
        discount_10_percent += 1
    elif clients % 5 == 0:
        coupon_200_rub += 1

print(clients, revenue, discount_10_percent, coupon_200_rub, teddy_bear)
````

#### Задача 10

Антонио открыл пиццерию и хочет привлечь новых клиентов. Для этого он придумал акцию: клиент получает две маленькие пиццы, которые в сумме будут большей площади, чем одна большая пицца конкурентов.

Пиццы конкурентов имеют различные размеры. Для того, чтобы предложение Антонио было привлекательным для клиентов, он хочет быстро определять, какой минимальный диаметр должны иметь две маленькие пиццы, чтобы их общая площадь была больше, чем площадь одной большой пиццы.

Напиши программу, которая на вход принимает диаметр пиццы конкурента и рассчитывает диаметр маленьких пицц.

Программа должна вернуть последовательно 3 числа:
- минимальный размер маленький пиццы,
- площадь двух маленьких пицц,
- площадь одной большой пиццы.

Диаметр пицц измеряется в сантиметрах и может быть только целым числом.

Пример ввода:
```
45
```

Ожидаемый результат:
```
32
1607.68
1589.625
```

````python
# Твой код здесь

import math

def calculate_pizza_sizes(competitor_diameter):
    competitor_radius = competitor_diameter / 2
    competitor_area = math.pi * (competitor_radius ** 2)
    
    small_diameter = 1
    while True:
        small_radius = small_diameter / 2
        small_area = math.pi * (small_radius ** 2)
        if 2 * small_area > competitor_area:
            break
        small_diameter += 1
    
    return small_diameter, 2 * small_area, competitor_area

competitor_diameter = int(input())
small_diameter, small_pizzas_area, competitor_pizza_area = calculate_pizza_sizes(competitor_diameter)

print(small_diameter)
print(f"{small_pizzas_area:.2f}")
print(f"{competitor_pizza_area:.3f}")
````

Владимир работает аналитиком данных исторических событий. В этом году он проводит ретроспективное исследование основных культурных событий.

У Владимира есть гипотеза: он считает, что важнейшие культурные события проводятся в високосный год, и количество високосных лет в насыщенные исторические периоды будет больше, чем в другие.

Напиши программу, которая принимает на вход год начала (в формате `ГГГГ`) и год окончания (в формате `ГГГГ`) исторического периода, а выводит количество високосных лет в этом периоде в виде целого числа.

**Для справки.** Год является високосным, если:
- делится на 400 или
- делится на 4 и не делится на 100.

```
2000

```

Ожидаемый результат:

````python
# Твой код здесь

def is_leap_year(year):
    return year % 400 == 0 or (year % 4 == 0 and year % 100 != 0)

start_year = int(input())
end_year = int(input())

leap_years_count = sum(1 for year in range(start_year, end_year) if is_leap_year(year))

print(leap_years_count)
````

Алексей работает в службе информационной безопасности онлайн-банка и проектирует функцию двухфакторной аутентификации.

 <!-- После ввода правильного логина и пароля пользователю отправляется SMS-сообщение с четырехзначным кодом, который необходимо ввести для подтверждения личности. -->

Задача Алексея — написать алгоритм генерации четырёхзначных кодов. У него есть гипотеза, что много кодов с одинаковой суммой цифр могут стать  источником потенциальной уязвимости. Чтобы это проверить, нужна программа, которая рассчитывает, сколько существует различных четырёхзначных кодов с суммой цифр, равной заданному числу.

>**Примечание.** На программном уровне установлен запрет формировать код, где сумма цифр равна 0.

Напиши программу, которая запрашивает у пользователя число, проверяет его на валидность и возвращает в виде целого числа количество возможных четырёхзначных кодов, сумма цифр которых равна этому числу.


````python
# Твой код здесь
from itertools import product

def count_codes_with_sum(target_sum):
    count = 0
    for i in product(range(10), repeat=4):
        digits_sum = sum(i)
        if digits_sum == target_sum:
            count += 1
    return count

target_sum = int(input())

if target_sum > 0:
    print(count_codes_with_sum(target_sum))
````

#### Задача 15

Владимир разрабатывает программное обеспечение для анализа финансовых рынков. Чтобы оценить риски инвестиционных проектов, ему нужно решать квадратные уравнения вида

$\text{a}x^2+bx+c=0,$

где $a$, $b$ и $c$ — коэффициенты, определяемые по  экономическим показателям.

Чтобы оценить жизнеспособность проекта, Владимиру нужно вычислить квадратный корень из дискриминанта этого уравнения, чтобы дальше использовать эту величину в расчётах.

Проблема в том, что система, которая оценивает показатели проекта, разрабатывалась давно, и не умеет рассчитывать квадратные корни, но может выполнять простые математические операции.

Владимир решил воспользоваться методом Ньютона для нахождения квадратного корня числа. Этот метод заключается в последовательном вычислении приближений $x_n$ к корню числа $S$ по формуле:

$x_{n+1} = \frac{1}{2} \left(x_n + \frac{S}{x_n}\right)$,

где $x_n$ — текущее приближение, а $S$ — число, квадратный корень которого необходимо вычислить. Приемлемая точность для Владимира — $0,001$ (то есть нужно прекращать вычисления, если разница между текущим и предыдущим приближением становится меньше $0,001$).

- значение дискриминанта,
- приближенное значение его квадратного корня при помощи метода Ньютона.

```

```

````python
# Твой код здесь

a = float(input())
b = float(input())
c = float(input())

D = b**2 - 4*a*c

def sqrt(s):
    x = s
    while True:
        _x = 0.5 * (x + s / x)
        if abs(x - _x) < 0.001:
            return _x
        x = _x

sqrt_D = sqrt(D)
print(D)
print(sqrt_D)
````

#### Задача 16

Сергей разрабатывает дисплейные системы для носимой электроники. Последняя задача, над которой он трудится — создать эффект «бегущего огня» на светодиодной панели. В самом простом случае это последовательное перемещение активного светодиода от одного края панели к другому.

Дисплей можно представить в виде одномерной горизонтальной строки, состоящей из `n` выключенных светодиодов, обозначаемых цифрой `0`. В крайней правой позиции расположен активный светодиод, который отмечен цифрой `1`.

Тогда движение огня для панели из 6 светодиодов будет выглядеть так:
```
000001
000010
000100
....
100000
```

Напиши программу, которая принимает на вход размер светодиодной панели и моделирует описанный визуальный эффект.

Пример ввода:
```
3
```

Ожидаемый результат:
```
001
010
100
```

````python
# Твой код здесь

n = int(input())

for i in range(n):
    panel = ['0'] * n
    panel[n - 1 - i] = '1'
    print(''.join(panel))
````
