>**Замечание.** Для простоты и массу, и рост выводи целыми числами.


## Задача 6. Подсчёт суммы покупок

Напиши функцию, которая будет помогать пользователю отслеживать, сколько денег он потратил на все покупки.

Программа должна использовать цикл `while` и при каждой покупке запрашивать у пользователя, сколько денег он потратил, а затем эту сумму прибавлять к общему количеству израсходованных денег.

Когда пользователь введет `0`, программа должна вывести общую потраченную сумму.

````python
# Напиши код здесь

total = 0
while True:
    purchase = int(input('Введите сумму покупки: '))
    total += purchase
    if purchase == 0:
        break

print(f'Общая потраченная сумма: {total}')
````

## Задача 7. Сколько осталось до дедлайна?

Виртуальный помощник помогает пользователю следить за прогрессом выполнения задач.

Напиши программу, которая использует функции `range()` и `reversed()` для обратного отсчёта времени и выводит сообщение:
```
До завершения задачи осталось [n] дней
```


````python
# Напиши код здесь

days = int(input('Введите количество дней до дедлайна: '))

def choose_day_declension(number):
    if 11 <= number % 100 <= 14:
        return "дней"
    elif number % 10 == 1:
        return "день"
    elif 2 <= number % 10 <= 4:
        return "дня"
    else:
        return "дней"

for i in reversed(range(1, days + 1)):
    print(f'До завершения задачи осталось {i} {choose_day_declension(i)}')
    
````

## Задача 8. Определение погодных условий


В зависимости от погоды помощник должен сообщать пользователю, как ему стоит одеться:
* если температура ниже `0` — `Оденьтесь тепло`,
* от `0` до `10` — `Наденьте куртку`,
* выше `10` — `Лёгкая одежда подойдет`.


````python
# Напиши код здесь

temperature = int(input('Введите температуру: '))

if temperature < 0:
    print('Оденьтесь тепло')
elif 0 <= temperature <= 10:
    print('Наденьте куртку')
else:
    print('Лёгкая одежда подойдет')
````

## Задача 9. Схема тренировок


Считается, что заниматься через день — оптимальная схема тренировок. Поэтому для любого числа месяца помощник должен предлагать начать тренировки сегодня же и выдать список чисел месяца, в которые ещё предстоит заниматься.

Напиши программу, которая запрашивает у пользователя текущий день месяца и использует функцию `range()`, чтобы перечислить будущие числа для тренировок:

```
Сегодня ваш день для тренировки
Вам предстоит тренироваться в следующие дни: [числа]
```

Для простоты считай, что в любом месяце 30 дней.

````python
# Напиши код здесь

day = int(input('Введите текущий день месяца: '))
days = [str(d) for d in range(day + 2, 31, 2)]

print('Сегодня ваш день для тренировки')
print(f'Вам предстоит тренироваться в следующие дни: {', '.join(days)}.')
````

## Задача 10. Проверка погоды для занятий спортом

Напиши программу, которая принимает на вход два значения: температуру и влажность воздуха. Помощник должен рекомендовать занятия спортом на улице, если температура находится в диапазоне от 15 до 25 градусов Цельсия, а влажность не превышает 60%.

Если хотя бы одно из условий не выполнено, ассистент должен выводить сообщение:

```
Сегодня лучше позаниматься в помещении
```


````python
# Напиши код здесь

temperature = int(input('Введите температуру: '))
humidity = int(input('Введите влажность: '))

if 15 <= temperature <= 25 and humidity <= 60:
    print('Рекомендую позаниматься на улице')
else:
    print('Сегодня лучше позаниматься в помещении')
````
